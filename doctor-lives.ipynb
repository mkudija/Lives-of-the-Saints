```python
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as dt


df = pd.read_csv('doctors.csv')
df.head()
```

```python
%matplotlib inline

fig, ax = plt.subplots(figsize=(10,10))
ax = plt.hlines(df.Number, df.Birth, df.Death, colors='#2e5fac', linewidth=10)
plt.xlabel('Year')

# remove axes/spines
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['top'].set_position(('data',.5))
plt.gca().spines['right'].set_position(('data',0))
plt.gca().spines['left'].set_position(('data',0))
ax = plt.gca()
ax.yaxis.set_visible(False)

for item in range(0,len(df)):
    x = df.iloc[item,4] + 20 # canonization
    y = df.iloc[item,3] - .25
    s = df['Name'][item]
    plt.text(x, y, s, fontdict=None)
    plt.plot(x-20, y+.25, 'ro')

# legend
# ax = plt.hlines(88, 40, 200, colors='#2e5fac', linewidth=10)
# plt.text(220,37.75,'Life')
# plt.plot(180,37,'ro')
# plt.text(220,36.75,'Canonization')

# top axis
ytop = 38
plt.plot([0,2500],[ytop,ytop],'k-')
plt.plot([0,0],[ytop,ytop-.5],'k-')
plt.plot([500,500],[ytop,ytop-.5],'k-')
plt.plot([1000,1000],[ytop,ytop-.5],'k-')
plt.plot([1500,1500],[ytop,ytop-.5],'k-')
plt.plot([2000,2000],[ytop,ytop-.5],'k-')
plt.plot([2495,2495],[ytop,ytop-.5],'k-')

plt.text(0,ytop+.25,'0')
plt.text(460,ytop+.25,'500')
plt.text(950,ytop+.25,'1000')
plt.text(1445,ytop+.25,'1500')
plt.text(1945,ytop+.25,'2000')
plt.text(2445,ytop+.25,'2500')
plt.text(1000,ytop+2,'Doctors of the Church',fontsize='large')
plt.text(0,-4.5,'The 35 Doctors of the Church by Christopher Rengers',fontsize=9)

plt.axvline(x=2020, linestyle=':', linewidth=.5, color='#64666a')

plt.savefig('doctors.png', dpi=300)
```
